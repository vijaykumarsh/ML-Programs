```python
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import datasets
```

```python
boston=datasets.load_boston()
```

```python
boston
```

```python
import numpy 
```

```python
print("data: ", boston.keys())
print("number of samples x and number of features: ",
boston.data.shape
)
print("feature names: ", boston.feature_names)
print("target feature range: %.1f to %.1f " % (
numpy.min(boston.target), numpy.max(boston.target)
))
```

```python
# the relation btw each feature and price
fig=plt.figure(figsize=(9,9))
fig.subplots_adjust(left=0,right=1,bottom=0,top=1,hspace=0.5,wspace=0.5)
for i in range(len(boston.feature_names)):
    boston_x=boston.data[:, i]
    ax=fig.add_subplot(4,4,i+1,xlabel=boston.feature_names[i])
    ax.scatter(boston_x,boston.target,alpha=0.75)
plt.show()
```
